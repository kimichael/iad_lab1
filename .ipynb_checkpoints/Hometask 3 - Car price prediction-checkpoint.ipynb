```python
num_feats = ['year', 'tax', 'mileage', 'mpg', 'engineSize']
num_pipe = Pipeline(steps=[('scaler', StandardScaler())])

string_feats = ['model', 'transmission', 'fuelType']
string_encoder = OneHotEncoder(handle_unknown='ignore')

preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_pipe, num_feats),
        ('cat', string_encoder, string_feats)])
```

```python
clf = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', LinearRegression())])
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=10)
clf.fit(X_train, Y_train)
print("model score: %.3f" % clf.score(X_test, Y_test))
from sklearn.metrics import mean_squared_error as MSE

y_pred = clf.predict(X_test)
print(np.sqrt(MSE(Y_test, y_pred)))
```

```python
clf = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', RidgeCV())])
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=10)
clf.fit(X_train, Y_train)
print("model score: %.3f" % clf.score(X_test, Y_test))
from sklearn.metrics import mean_squared_error as MSE

y_pred = clf.predict(X_test)
print(np.sqrt(MSE(Y_test, y_pred)))
```

```python
object_cols = auto_data.columns[auto_data.dtypes == 'object']
le = LabelEncoder()
for col in object_cols:
    auto_data[col] = le.fit_transform(auto_data[col])
auto_data
```

```python
X, Y = auto_data.drop(['price'], axis=1), auto_data['price']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=10)
std_scaler = StandardScaler()
X_train_scaled = std_scaler.fit_transform(X_train)
X_test_scaled = std_scaler.transform(X_test)
X_train_scaled
```

```python
#Линейная регрессия
simple_LR = LinearRegression()
simple_LR.fit(X_train_scaled, Y_train)
# Строим прогноз
simple_LR_predictions = simple_LR.predict(X_test_scaled)
simple_LR_predictions
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(simple_LR_predictions, Y_test)
plt.xlabel('Predictions')
plt.ylabel('Target')
```

```python
simple_LR_RMSE = metrics.mean_squared_error(Y_test, simple_LR_predictions, squared = False)

print (f'RMSE = {round(simple_LR_RMSE,3)}')
```

```python
auto_data.describe()
```

```python
lasso = Lasso()
lasso.fit(X_train_scaled, Y_train)
lasso_predictions = lasso.predict(X_test_scaled)
lasso_predictions
```

```python
print (f'RMSE = {round(metrics.mean_squared_error(Y_test, lasso_predictions, squared=False), 3)}')
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(lasso_predictions, Y_test)
plt.xlabel('Predictions')
plt.ylabel('Target')
```

```python
lassoCV = LassoCV()
lassoCV.fit(X_train_scaled, Y_train)
lassoCV_predictions = lassoCV.predict(X_test_scaled)
lassoCV_predictions
print (f'RMSE = {round(metrics.mean_squared_error(Y_test, lasso_predictions, squared=False), 3)}')
plt.figure(figsize=(8, 8))
plt.scatter(lasso_predictions, Y_test)
plt.xlabel('Predictions')
plt.ylabel('Target')
```

```python
ridgeCV = RidgeCV()
ridgeCV.fit(X_train_scaled, Y_train)
ridgeCV_predictions = lassoCV.predict(X_test_scaled)
ridgeCV_predictions
print (f'RMSE = {round(metrics.mean_squared_error(Y_test, ridgeCV_predictions, squared=False), 3)}')
plt.figure(figsize=(8, 8))
plt.scatter(ridgeCV_predictions, Y_test)
plt.xlabel('Predictions')
plt.ylabel('Target')
```
